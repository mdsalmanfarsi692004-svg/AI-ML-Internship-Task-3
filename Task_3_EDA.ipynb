```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the Dataset
# Loading the standard Iris dataset directly from Seaborn
df = sns.load_dataset('iris')

print("--- DATASET PREVIEW ---")
display(df.head())
print(f"\nShape: {df.shape}")

# 2. Statistical Summary
print("\n--- STATISTICAL SUMMARY ---")
display(df.describe())

# 3. Check for Null Values
print("\n--- MISSING VALUES ---")
print(df.isnull().sum())

# --- VISUALIZATIONS ---

# 4. Numerical Distribution (Histograms)
plt.figure(figsize=(12, 6))
plt.suptitle("Distribution of Numerical Features", fontsize=16)
for i, col in enumerate(df.columns[:-1]): # Exclude target column 'species'
    plt.subplot(2, 2, i+1)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()

# 5. Categorical Feature Analysis (Count Plot)
plt.figure(figsize=(6, 4))
sns.countplot(x='species', data=df, palette='viridis')
plt.title("Class Balance (Target Variable)")
plt.show()

# 6. Outlier Detection (Box Plots)
plt.figure(figsize=(12, 6))
plt.suptitle("Box Plots to Detect Outliers", fontsize=16)
for i, col in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=col, data=df, palette='Set2')
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()

# 7. Correlation Heatmap
# Encode species to numeric to include in correlation or just drop it
numerical_df = df.drop(columns=['species'])
plt.figure(figsize=(8, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()

# 8. Pairplot (Relationships)
sns.pairplot(df, hue='species', palette='husl')
plt.suptitle("Pairplot of Features by Species", y=1.02)
plt.show()
```
